```python
def build_model():
    
    preprocessor = ColumnTransformer(
    [
        (
        "cat_cols",
        OneHotEncoder(sparse=False, handle_unknown="ignore"),
        CAT_COLS,
        ),
        ("cont_cols", "passthrough", NUM_COLS),
    ]
    )
    
    model = XGBClassifier(
    max_depth=7,
    min_child_weight=5,
    n_estimators=50,
    colsample=0.8,
    )
    return Pipeline([("preprocessor", preprocessor), ("model", model)])
```
